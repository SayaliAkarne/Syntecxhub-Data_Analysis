```python
To compute and visualize correlations between numerical variables using a heatmap.
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("C:\\Users\\Dell\\Downloads\\sales_dataset.csv")

# Select only numerical columns
numeric_df = df.select_dtypes(include=['number'])

# Compute correlation
corr_matrix = numeric_df.corr()

# Plot heatmap
plt.figure(figsize=(8,6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.savefig("heatmap.png")
plt.show()
```

```python
# Pairplot for all numerical variables
sns.pairplot(numeric_df)

plt.savefig("pairplot.png")
plt.show()
```

```python
# Get correlation values
corr_matrix = numeric_df.corr()

# Print strong positive and negative correlations
print("Strong Positive Correlations (> 0.5):")
print(corr_matrix[corr_matrix > 0.5])

print("\nStrong Negative Correlations (< -0.5):")
print(corr_matrix[corr_matrix < -0.5])
```

```python
import pandas as pd

# Step 1: Load dataset
df = pd.read_csv("C:\\Users\\Dell\\Downloads\\sales_dataset.csv")

# Step 2: Select numerical columns
numeric_df = df.select_dtypes(include=['number'])

# Step 3: Create correlation matrix
corr_matrix = numeric_df.corr()

# Step 4: Convert to pairs
corr_pairs = corr_matrix.unstack()

# Step 5: Remove self correlations
corr_pairs = corr_pairs[corr_pairs.index.get_level_values(0) != corr_pairs.index.get_level_values(1)]

# Step 6: Strong positive correlations
strong_pos = corr_pairs[corr_pairs > 0.5]

print("Strong Positive Correlations:")
print(strong_pos.sort_values(ascending=False))

# Step 7: Strong negative correlations
strong_neg = corr_pairs[corr_pairs < -0.5]

print("\nStrong Negative Correlations:")
print(strong_neg)
```
